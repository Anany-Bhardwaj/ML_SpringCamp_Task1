```python
#AI generated

def generate_ppnf_features(split_data_cleaned):
    # Interaction terms: Pairwise product of features
    for i in range(1, 40):
        for j in range(i + 1, 40):
            split_data_cleaned[f'Feature_{i}_x_Feature_{j}'] = split_data_cleaned[f'Feature {i}'] * split_data_cleaned[f'Feature {j}']

    # Ratio-based features: High-to-low ratios
    for i in range(1, 40):
        for j in range(i + 1, 40):
            split_data_cleaned[f'Feature_{i}_div_Feature_{j}'] = np.where(
                split_data_cleaned[f'Feature {j}'] != 0,
                split_data_cleaned[f'Feature {i}'] / split_data_cleaned[f'Feature {j}'],
                0
            )

    # Normalized features: Features normalized to their sum
    for i in range(1, 40):
        split_data_cleaned[f'Feature_{i}_normalized'] = split_data_cleaned[f'Feature {i}'] / split_data_cleaned[[f'Feature {k}' for k in range(1, 40)]].sum(axis=1)

    # Aggregate statistical features
    split_data_cleaned['Feature_Mean'] = split_data_cleaned[[f'Feature {i}' for i in range(1, 40)]].mean(axis=1)
    split_data_cleaned['Feature_StdDev'] = split_data_cleaned[[f'Feature {i}' for i in range(1, 40)]].std(axis=1)
    split_data_cleaned['Feature_Max'] = split_data_cleaned[[f'Feature {i}' for i in range(1, 40)]].max(axis=1)
    split_data_cleaned['Feature_Min'] = split_data_cleaned[[f'Feature {i}' for i in range(1, 40)]].min(axis=1)

    # Peak-to-average power ratio (PAPF): Max feature divided by the mean
    split_data_cleaned['PAPF'] = split_data_cleaned['Feature_Max'] / split_data_cleaned['Feature_Mean']

    return split_data_cleaned

# Apply the function to the dataset
split_data_cleaned = generate_ppnf_features(split_data_cleaned)

# Display the resulting data with new features
split_data_cleaned.head()
```

```python
#AI
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix for PPNF and all other features
correlation_matrix = split_data_cleaned.corr()

# Extract correlations only for the PPNF column
ppnf_correlation = correlation_matrix['PPNF'].drop('PPNF')

# Sort the correlations in descending order
sorted_ppnf_correlation = ppnf_correlation.sort_values(ascending=False)

# Print the sorted correlation values
print(sorted_ppnf_correlation)

# Plot a heatmap for better visualization
plt.figure(figsize=(12, 8))
sns.heatmap(
    sorted_ppnf_correlation.to_frame(),
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    cbar=True
)
plt.title('Correlation of PPNF with Features')
plt.show()
```

```python
#AI
# Set a threshold for correlation
threshold = 0.2

# Select features with correlation > threshold (positive or negative)
selected_features = sorted_ppnf_correlation[abs(sorted_ppnf_correlation) > threshold].index

# Create X_ppnf with the selected features
X_ppnf = split_data_cleaned[selected_features]

# Display selected features
print("Selected Features for PPNF Training:")
print(selected_features)

# Check the shape of X_ppnf
print(f"Shape of X_ppnf: {X_ppnf.shape}")
```
